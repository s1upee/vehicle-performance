```python
# Load the anomaly detection results
df = pd.read_csv("data/optimized_ml_detected_anomalies.csv")

# Show basic info
print("Data Shape:", df.shape)
print("First 5 rows:")
display(df.head())  # Use display() in Jupyter for better formatting

# Check for missing values
print("Missing values:\n", df.isnull().sum())
```

```python
def plot_anomalies(feature_name, title, color):
    plt.figure(figsize=(12, 6))
    sns.lineplot(x=df.index, y=df[feature_name], label=feature_name, color=color)
    
    # Highlight anomalies
    anomalies = df[df["final_anomaly"] == 1]
    plt.scatter(anomalies.index, anomalies[feature_name], color="red", label="Anomaly", zorder=3)
    
    plt.title(title)
    plt.xlabel("Index")
    plt.ylabel(feature_name)
    plt.legend()
    plt.show()
```

```python
plot_anomalies("acceleration", "Vehicle Acceleration Over Time with Anomalies", "blue")
plot_anomalies("brake_force", "Braking Force Over Time with Anomalies", "orange")
plot_anomalies("steering_angle", "Steering Angle Over Time with Anomalies", "green")
```

```python
# Change directory if needed
os.chdir("/Users/elizavetakrasiukova/Desktop/kia-intern/test2")

# Load anomaly detection results
df = pd.read_csv("data/optimized_ml_detected_anomalies.csv")

# Display first few rows
df.head()
```

```python
if "final_anomaly" not in df.columns:
    raise ValueError("Column 'final_anomaly' not found in dataset. Ensure the correct file is being used.")

# Separate normal and anomalous data
normal_data = df[df["final_anomaly"] == 0]
anomalies = df[df["final_anomaly"] == 1]

# Display the count
print(f"Total data points: {len(df)}")
print(f"Normal instances: {len(normal_data)}")
print(f"Anomalies detected: {len(anomalies)}")
```

```python
def plot_anomalies(feature_name, title, color):
    plt.figure(figsize=(12, 6))
    sns.lineplot(x=df.index, y=df[feature_name], label=feature_name, color=color)
    plt.scatter(anomalies.index, anomalies[feature_name], color="red", label="Anomaly", zorder=3)
    plt.title(title)
    plt.xlabel("Index")
    plt.ylabel(feature_name)
    plt.legend()
    plt.show()

# Plot anomalies for key features
plot_anomalies("acceleration", "Vehicle Acceleration Over Time with Anomalies", 'blue')
plot_anomalies("brake_force", "Braking Force Over Time with Anomalies", 'orange')
plot_anomalies("steering_angle", "Steering Angle Over Time with Anomalies", 'green')
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[["acceleration", "brake_force", "steering_angle"]], showfliers=True)
plt.title("Distribution of Features with Anomalies")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Matrix")
plt.show()
```
